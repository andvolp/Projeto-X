```python
import pandas as pd
import numpy as np
import tensorflow
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization
import tensorflow as tf
import io
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.application import MIMEApplication
#from google.colab import files
```

```python

df = pd.read_csv('https://raw.githubusercontent.com/andvolp/Projeto-X/main/Projeto%201/winequality-red.csv', encoding='utf-8', sep=';')

dfs = np.array(df)

for i in range(len(dfs)):
  if dfs[i,11] > 5:
    dfs[i,11] = 1
  else:
    dfs[i,11] = 0

# utilizarei o keras

X = dfs[:,:-1]
y = dfs[:,11]

scaler =StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X , y, test_size=0.2, random_state=3)

model = Sequential()
#Input
model.add(tf.keras.Input(shape=(11,)))

#layers
#model.add(tf.keras.layers.Activation('tanh'))
#model.add(tf.keras.layers.Activation('tanh'))
#model.add(tf.keras.layers.Activation('tanh'))

#Fully layer 1
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(1024))
model.add(BatchNormalization())
model.add(tf.keras.layers.Activation('relu'))
model.add(tf.keras.layers.Dropout(0.2))

#Fully layer 2
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(1024))
#model.add(BatchNormalization())
model.add(tf.keras.layers.Activation('relu'))
model.add(tf.keras.layers.Dropout(0.2))

#Output
model.add(Dense(1))
model.add(tf.keras.layers.Activation('sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

historico = model.fit(X_train, y_train, epochs=10, batch_size=64, verbose=1)

loss, accuracy = model.evaluate(X_test, y_test, verbose=0)

log_stream = io.StringIO()
log_stream.write(f"Loss: {loss}, Accuracy: {accuracy}\n")
log_stream.write(str(historico.history))

 #Salvar o arquivo de log em disco
with open('historico.log', 'w') as f:
    f.write(log_stream.getvalue())

#files.download('/content/history.log')

print('Acuracia:', accuracy)
print('loss', loss)
```
