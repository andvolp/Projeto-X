```python
import pandas as pd
import numpy as np
import tensorflow
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization
import tensorflow as tf
import io
from tensorflow.keras.utils import to_categorical
#from google.colab import files
```

```python
def busca(X,y):
  scaler =StandardScaler()
  X = scaler.fit_transform(X)

  X_train, X_test, y_train, y_test = train_test_split(X , y, test_size=0.2, random_state=3)

  model = Sequential()

  model.add(tf.keras.Input(shape=(7,)))

  #Fully layer 1
  model.add(tf.keras.layers.Flatten())
  model.add(tf.keras.layers.Dense(1024))
  model.add(tf.keras.layers.Activation('relu'))
  model.add(tf.keras.layers.Dropout(0.1))

  #Fully layer 2
  model.add(tf.keras.layers.Flatten())
  model.add(tf.keras.layers.Dense(1024))
  #model.add(BatchNormalization())
  model.add(tf.keras.layers.Activation('relu'))
  model.add(tf.keras.layers.Dropout(0.1))

  #Output
  model.add(Dense(1))
  model.add(tf.keras.layers.Activation('sigmoid'))

  model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

  historico = model.fit(X_train, y_train, epochs=100, batch_size=64, verbose=1)

  loss, accuracy = model.evaluate(X_test, y_test, verbose=0)

  log_stream = io.StringIO()
  log_stream.write(f"Loss: {loss}, Accuracy: {accuracy}\n")
  log_stream.write(str(historico.history))

  #Salvar o arquivo de log em disco
  with open('historico.log', 'w') as f:
      f.write(log_stream.getvalue())

  #files.download('/content/history.log')

  print('Acuracia:', accuracy)
  print('loss', loss)
  return accuracy
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/andvolp/Projeto-X/main/Projeto%201/winequality-red.csv', encoding='utf-8', sep=';')

df = np.array(df)

for i in range(len(df)):
  if df[i,11] > 5:
    df[i,11] = 1
  else:
    df[i,11] = 0

X = df[:,:-1]
y = df[:,11]

#z = X[:,[0,10]]
#print(z)

acuracia_max = 0
melhores_colunas = ""
resultados_acuracia = {}

for i in range(X.shape[1]):
  for j in range(i + 1, X.shape[1]):
    for k in range(j + 1, X.shape[1]):
      for l in range(k + 1, X.shape[1]):
        for m in range(l + 1, X.shape[1]):
          for n in range(m + 1, X.shape[1]):
            for o in range(n + 1, X.shape[1]):
              z = X[:,[i,j,k,l,m,n,o]]
              acuracia = busca(z,y)
              colunas = "{}+{}+{}+{}+{}+{}+{}".format(i,j,k,l,m,n,o)
              resultados_acuracia[colunas] = acuracia

            if acuracia > acuracia_max:
              acuracia_max = acuracia
              melhores_colunas = colunas

print(acuracia_max)
print(melhores_colunas)
```
